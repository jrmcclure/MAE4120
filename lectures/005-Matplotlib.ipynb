### Example ###

Jupyter magic command `%matplotlib inline` ensures all matplotlib plots are shown inline

```python
%matplotlib inline
```

Import `numpy` under the name `np` and `matplotlib.pyplot` under the name `plt` to give access to library commands

```python
import numpy as np
import matplotlib.pyplot as plt
```

Using numpy's MATLAB style API generate (x,y) data

```python
x = np.linspace(0, 5, 100)
y = x ** 2
```

Generate a simple plot

```python
fig, ax = plt.subplots()     # Create figure
plt.plot(x, y, 'r')    # Plot x vs. y using color red ('r')
plt.xlabel('x')        # Set X label
plt.ylabel('y')        # Set Y label
plt.title('y = x^2')   # Set plot title
plt.show()             # Display the plot
```

### 2 plots on 1 Figure ###
---
Generate a new set of (x,z) data

```python
z = 25 - x ** 2
```

Plot both sets of data on the sample figure

```python
fig, ax = plt.subplots()                 # Create figure
plt.plot(x, y, 'b--', x, z, 'g')   # Plot y(x) & z(x)
plt.xlabel('x')                    # Set X label
plt.ylabel('y / z')                # Set Y label
plt.title('y = x^2, z = 25 - x^2') # Set plot title
plt.show()                         # Display the plot   
```

### Plotting Lines Point-to-Point ###
---

```python
# Line 1
x1 = (0,1)   # x-start, x-end
y1 = (0,1)   # y-start, y-end
# Line 2
x2 = (0,3)
y2 = (1,.5)
# Line 3
x3 = (2,0)
y3 = (.1,.8)
# Line 4
x4 = (2,3)
y4 = (1,0)

fig, ax = plt.subplots()         # Create figure
plt.plot(x1, y1, color='b', marker='o')                      # Plot Line 1
plt.plot(x2, y2, color='g', marker='o')                      # Plot Line 2
plt.plot(x3, y3, color='purple', marker='o')                 # Plot Line 3
plt.plot(x4, y4, color='red', marker='o', linestyle='--')    # Plot Line 4
plt.show()                 # Display the plot   
```

### Plotting Linkages ###
---

Import `math` module

```python
import math
```

Inputs for the solved linkage.

> $a,b,c,d,\theta_{1}$ are the inputs to our `fourbar_calc_theta3` funtion.

> $\theta_{3}$ is the output of our `fourbar_calc_theta3` function.

```python
a,b,c,d = 200,400,400,800
theta1 = 30
theta3 = 133.44
```

Solve for the endpoints of Link 1

```python
L1_x0 = 0
L1_x1 = a * math.cos(math.radians(theta1))
L1x = (L1_x0,L1_x1)

L1_y0 = 0
L1_y1 = a * math.sin(math.radians(theta1))
L1y = (L1_y0,L1_y1)
```

Solve for the endpoints of Link 4

```python
L4_x0 = 0
L4_x1 = d
L4x = (L4_x0,L4_x1)

L4_y0 = 0
L4_y1 = 0
L4y = (L4_y0,L4_y1)
```

Solve for the endpoints of Link 3

```python
L3_x0 = L4_x1
L3_x1 = c * math.cos(math.radians(theta3))
L3x = (L3_x0,L3_x1)

L3_y0 = L4_y1
L3_y1 = c * math.sin(math.radians(theta3))
L3y = (L3_y0,L3_y1)
```

Solve for the endpoints of Link 2

```python
L2_x0 = L1_x1
L2_x1 = L3_x1
L2x = (L2_x0,L2_x1)

L2_y0 = L1_y1
L2_y1 = L3_y1
L2y = (L2_y0,L2_y1)
```

Now that we have all the endpoints, we can plot the linkage just as we did above

```python
# Create figure and axes
fig, ax = plt.subplots(figsize=(8,8))

# Add data to the plot
plt.plot(L1x, L1y, color='b', marker='o')                      # Plot Line 1
plt.plot(L2x, L2y, color='g', marker='o')                      # Plot Line 2
plt.plot(L3x, L3y, color='purple', marker='o')                 # Plot Line 3
plt.plot(L4x, L4y, color='red', marker='o', linestyle='--')    # Plot Line 4
plt.show()
```

__Problem:__ Axes have different scales, distorting the image

### Better Linkage Plots ###
 - Make both axes use the same scale
 - Set x & y-axis limits
 - Set x & y-axis tick marks
 - Add grid lines

```python
# Create figure and axes
fig, ax = plt.subplots(figsize=(12,12))         

# Set axes to both use same scale
ax.set_aspect('equal')

# Set min & max limits for both axes
ax.set_xlim(( -300, 900))
ax.set_ylim(( -50, 350))

# Set tick marks for both axes
ax.xaxis.set_ticks(np.arange(-300, 901, 100))
ax.yaxis.set_ticks(np.arange(0, 301, 100))

# Set grid lines
ax.grid(b=True, which='major', color='gray', linestyle='--')

# Add data to the plot
plt.plot(L1x, L1y, color='b', marker='o')
plt.plot(L2x, L2y, color='g', marker='o')
plt.plot(L3x, L3y, color='purple', marker='o')
plt.plot(L4x, L4y, color='red', marker='o', linestyle='--')
plt.show()
```
